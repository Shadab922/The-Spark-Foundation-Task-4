# Data Visualization

```python
df.hist(figsize=(50,30))
plt.show()
```

# Numbers of Attacks from 1970 to 2017

```python
plt.subplots(figsize=(20,10))
sns.countplot(x=df['iyear'])
plt.xticks(rotation=70)
plt.title('Terrorist attacks from 1970 to 2017')
plt.show()
```

# Types of Attacks by Terrorist

```python
plt.figure(figsize=(10,10))
sns.countplot(x=df['attacktype1_txt'],order=df['attacktype1_txt'].value_counts().index)
plt.xticks(rotation=70)
plt.xlabel('Attack types')
plt.ylabel('No. of attacks')
plt.show()
```

# Attack Targets of Terrorists

```python
plt.figure(figsize=(20,10))
sns.countplot(x=df['targtype1_txt'],order=df['targtype1_txt'].value_counts().index)
plt.xticks(rotation=70)
plt.xlabel('Targets')
plt.ylabel('No. of Targets')
plt.show()
```

# Names of Terrorist Groups

```python
plt.figure(figsize=(15,6))
sns.barplot(x=df['gname'].value_counts()[:20].index, y=df['gname'].value_counts()[:20].values)
plt.xticks(rotation=70)
plt.title('Total Terrorist Groups')
plt.xlabel('Group Names')
plt.ylabel('Total no. of Attacks')
plt.show()
```

# Weapons Used in Terrorist Attacks

```python
df['weaptype1_txt'].value_counts()
plt.figure(figsize=(20,10))
sns.barplot(x=df['weaptype1_txt'].value_counts()[:20].index,y=df['weaptype1_txt'].value_counts()[:20].values)
plt.title('Weapon Usage')
plt.xticks(rotation=70)
plt.xlabel('Weapon Used')
plt.ylabel('No. of Times')
plt.show()
```

# Region Affected by Terrorist

```python
df['region_txt'].value_counts()
plt.figure(figsize=(20,10))
sns.countplot(x=df['region_txt'])
plt.xticks(rotation=70)
plt.show()
```

# Cities Affected by Terrorist

```python
df['city'].value_counts()
plt.figure(figsize=(20,10))
sns.barplot(x=df['city'].value_counts()[:20].index,y=df['city'].value_counts()[:20].values)
plt.title('Cities got Attacked')
plt.xticks(rotation=70)
plt.xlabel('Cities')
plt.ylabel('No. of attacks')
plt.show()
```

# Countries Affected by Terrorist Attacks

```python
df['country'].value_counts()
plt.figure(figsize=(20,10))
sns.barplot(x=df['country_txt'].value_counts()[:20].index,y=df['country_txt'].value_counts()[:20].values)
plt.title('Countries got Attacked')
plt.xticks(rotation=70)
plt.xlabel('Countries')
plt.ylabel('No. of Attacks')
plt.show()
```
